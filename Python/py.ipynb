```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("books.csv")
smallfont = {"color": "red", "size": 15}
x = [203,178,192,562,152,89,293,144,189,103]
y = ["Seas Bay","Crusty Crab","Manderin","Jurrasic Restaurant","All you can eat Joes","Tandoori Flame","Gulab Jamun Wale","The Keg","Incredible India","Fionn Maccools"]
plt.figure(figsize=[10,10])
plt.pie(x, labels=y, autopct='%1.1f%%',  startangle=90, explode=[0,0,0,0.25,0,0,0,0,0,0])
plt.xlabel("Restaurant", fontdict=smallfont)
plt.ylabel("Amount of people", fontdict=smallfont)
plt.title("Amounts of people daily in the restaurant average")
plt.show()
```
